```python
import pandas as pd

# Crear DataFrame resumen con totales de cada tienda
totales = pd.DataFrame({
    'Tienda': ['Tienda', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ventas': [
        tienda['Precio'].sum(),
        tienda2['Precio'].sum(),
        tienda3['Precio'].sum(),
        tienda4['Precio'].sum()
    ]
})

# Calcular el total general
total_general = totales['Ventas'].sum()

# Calcular el porcentaje de participación
totales['% Participación'] = (totales['Ventas'] / total_general * 100).round(2)

# Agregar columna formateada en pesos
totales['Ventas_formateadas'] = totales['Ventas'].apply(
    lambda x: f"${x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
)

# Ordenar de menor a mayor venta (para ver la menos comercial primero)
totales = totales.sort_values(by='Ventas', ascending=True)

print(totales)
```

```python
import matplotlib.pyplot as plt

# Configurar el tamaño del gráfico
plt.figure(figsize=(8,5))

# Crear gráfico de barras horizontal (ordenado de menor a mayor)
plt.barh(totales['Tienda'], totales['Ventas'], color='orange')

# Agregar títulos y etiquetas
plt.title('Ventas por Tienda', fontsize=14, fontweight='bold')
plt.xlabel('Ventas ($)', fontsize=12)
plt.ylabel('Tienda', fontsize=12)

# Mostrar los valores al lado de cada barra
for index, value in enumerate(totales['Ventas']):
    plt.text(value + (value*0.01), index, f"${value:,.0f}".replace(",", "."),
             va='center', fontsize=10)


plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(6,6))

plt.pie(
    totales['Ventas'],
    labels=totales['Tienda'],
    autopct='%1.1f%%',        # muestra el porcentaje con 1 decimal
    startangle=90,            # empieza desde arriba
    colors=plt.cm.Paired.colors,  # paleta de colores bonita
    wedgeprops={'edgecolor': 'white'}  # bordes blancos entre sectores
)

plt.title('Participación de Ventas por Tienda', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1= pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

t1 = pd.read_csv(url)
t2 = pd.read_csv(url2)
t3 = pd.read_csv(url3)
t4 = pd.read_csv(url4)

# --- Convertir la fecha ---
for df in [t1, t2, t3, t4]:
    df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'], dayfirst=True)
    df['Total'] = df['Precio'] + df['Costo de envío']

# --- Agrupar por fecha ---
t1_agg = t1.groupby('Fecha de Compra')['Total'].sum().reset_index().rename(columns={'Total':'Tienda1'})
t2_agg = t2.groupby('Fecha de Compra')['Total'].sum().reset_index().rename(columns={'Total':'Tienda2'})
t3_agg = t3.groupby('Fecha de Compra')['Total'].sum().reset_index().rename(columns={'Total':'Tienda3'})
t4_agg = t4.groupby('Fecha de Compra')['Total'].sum().reset_index().rename(columns={'Total':'Tienda4'})

# --- Unir por fecha ---
ventas_total = t1_agg.merge(t2_agg, on='Fecha de Compra', how='outer') \
                     .merge(t3_agg, on='Fecha de Compra', how='outer') \
                     .merge(t4_agg, on='Fecha de Compra', how='outer') \
                     .sort_values('Fecha de Compra')

# --- Graficar ---
plt.figure(figsize=(12,7))
plt.plot(ventas_total['Fecha de Compra'], ventas_total['Tienda1'], marker='o', label='Tienda 1')
plt.plot(ventas_total['Fecha de Compra'], ventas_total['Tienda2'], marker='o', label='Tienda 2')
plt.plot(ventas_total['Fecha de Compra'], ventas_total['Tienda3'], marker='o', label='Tienda 3')
plt.plot(ventas_total['Fecha de Compra'], ventas_total['Tienda4'], marker='o', label='Tienda 4')

plt.title(' grafico3 : Evolución de Ventas por Tienda en el Tiempo')
plt.xlabel('Fecha de Compra')
plt.ylabel('Total de Ventas')
plt.legend()
plt.grid(True)
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Crear una nueva columna con el año
for df in [t1, t2, t3, t4]:
    df['Año'] = pd.to_datetime(df['Fecha de Compra'], dayfirst=True).dt.year
    df['Total'] = df['Precio'] + df['Costo de envío']

# Agrupar por año y tienda
t1_agg = t1.groupby('Año')['Total'].sum().reset_index().rename(columns={'Total': 'Tienda1'})
t2_agg = t2.groupby('Año')['Total'].sum().reset_index().rename(columns={'Total': 'Tienda2'})
t3_agg = t3.groupby('Año')['Total'].sum().reset_index().rename(columns={'Total': 'Tienda3'})
t4_agg = t4.groupby('Año')['Total'].sum().reset_index().rename(columns={'Total': 'Tienda4'})

# Unir todas las tiendas
ventas_anual = t1_agg.merge(t2_agg, on='Año', how='outer') \
                     .merge(t3_agg, on='Año', how='outer') \
                     .merge(t4_agg, on='Año', how='outer')

# Graficar
plt.figure(figsize=(10,6))
plt.plot(ventas_anual['Año'], ventas_anual['Tienda1'], marker='o', label='Tienda 1')
plt.plot(ventas_anual['Año'], ventas_anual['Tienda2'], marker='o', label='Tienda 2')
plt.plot(ventas_anual['Año'], ventas_anual['Tienda3'], marker='o', label='Tienda 3')
plt.plot(ventas_anual['Año'], ventas_anual['Tienda4'], marker='o', label='Tienda 4')

plt.title('Grafico4: Evolución Anual de Ventas por Tienda')
plt.xlabel('Año')
plt.ylabel('Total de Ventas')
plt.legend()
plt.grid(True)
plt.show()
```
