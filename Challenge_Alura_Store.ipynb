```python
import pandas as pd

# Crear DataFrame resumen con totales de cada tienda
totales = pd.DataFrame({
    'Tienda': ['Tienda', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ventas': [
        tienda['Precio'].sum(),
        tienda2['Precio'].sum(),
        tienda3['Precio'].sum(),
        tienda4['Precio'].sum()
    ]
})

# Calcular el total general
total_general = totales['Ventas'].sum()

# Calcular el porcentaje de participación
totales['% Participación'] = (totales['Ventas'] / total_general * 100).round(2)

# Agregar columna formateada en pesos
totales['Ventas_formateadas'] = totales['Ventas'].apply(
    lambda x: f"${x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
)

# Ordenar de menor a mayor venta (para ver la menos comercial primero)
totales = totales.sort_values(by='Ventas', ascending=True)

print(totales)
```

```python
import matplotlib.pyplot as plt

# Configurar el tamaño del gráfico
plt.figure(figsize=(8,5))

# Crear gráfico de barras horizontal (ordenado de menor a mayor)
plt.barh(totales['Tienda'], totales['Ventas'], color='orange')

# Agregar títulos y etiquetas
plt.title('Ventas por Tienda', fontsize=14, fontweight='bold')
plt.xlabel('Ventas ($)', fontsize=12)
plt.ylabel('Tienda', fontsize=12)

# Mostrar los valores al lado de cada barra
for index, value in enumerate(totales['Ventas']):
    plt.text(value + (value*0.01), index, f"${value:,.0f}".replace(",", "."),
             va='center', fontsize=10)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(6,6))

plt.pie(
    totales['Ventas'],
    labels=totales['Tienda'],
    autopct='%1.1f%%',        # muestra el porcentaje con 1 decimal
    startangle=90,            # empieza desde arriba
    colors=plt.cm.Paired.colors,  # paleta de colores bonita
    wedgeprops={'edgecolor': 'white'}  # bordes blancos entre sectores
)

plt.title('Participación de Ventas por Tienda', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd


tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

todas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)


comparacion = todas.groupby('Tienda').agg(
    Ventas_Totales = ('Precio', 'sum'),
    Ticket_Promedio = ('Precio', 'mean'),
    Cantidad_Ventas = ('Precio', 'count'),
    Costo_Promedio_Envio = ('Costo de envío', 'mean')
).round(2)

comparacion

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ventas Totales
plt.figure()
comparacion['Ventas_Totales'].plot(kind='bar')
plt.xlabel('Tienda')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por Tienda')
plt.show()

# Ticket Promedio
plt.figure()
comparacion['Ticket_Promedio'].plot(kind='barh', color='skyblue')
plt.xlabel('Tienda')
plt.ylabel('Ticket Promedio')
plt.title('Ticket Promedio por Tienda')
plt.show()

# Costo Promedio de Envío
plt.figure()
comparacion['Costo_Promedio_Envio'].plot(kind='bar')
plt.xlabel('Tienda')
plt.ylabel('Costo Promedio de Envío')
plt.title('Costo Promedio de Envío por Tienda')
plt.show()
```

##**Se observa en los graficos de barras que la Tienda 4 presenta las menores Ventas Totales y el Ticket Promedio más bajo, lo que indica una menor capacidad de generar ingresos por cliente.**
